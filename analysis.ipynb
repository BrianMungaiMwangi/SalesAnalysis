```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
#opening the file.
data = pd.read_csv("Sample-Superstore.csv")
data
```

```python
#Statistical Summary
data.describe()
```

```python
#Interpretive text
summary = data[['Ship Mode', 'Segment', 'City', 'Sales', 'Quantity', 'Discount', 'Profit']]
summary
```

```python
data['Ship Mode'].value_counts()
```

```python
#Visualised  Ship Mode counts
ship_mode = ['Standard Class', 'First Class', 'Second Class', 'Same Day']
value_count = data['Ship Mode'].value_counts()
plt.plot(ship_mode, value_count)
plt.title("Value Count of Ship Mode")
plt.xlabel("Ship Mode")
plt.ylabel("Value Count")
plt.yticks(np.arange(500, 6500, 500))
plt.tight_layout()
plt.show()
```

```python
#Create a subset
subset = data[['Quantity', 'City', 'Sales']]
subset
```

```python
#Subset statistics
subset.describe()
```

```python
#Grouping data in a subset.
df = pd.DataFrame(subset)
subset_data = df.groupby(['City', 'Quantity'])
for (city, quantity), group in subset_data:
    print(f"City: {city}, Quantity: {quantity}\n{group}\n")


```

```python
#demonstrate how this split impacts the distribution of Sales.
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))

for (quantity, city), ax in zip(df.groupby(['Quantity', 'City']), axes.flat):
    ax.hist(city['Sales'], bins=10)
    ax.set_title(f"Quantity {quantity[0]}, City {quantity[1]}")
    ax.set_xlabel("Sales")
    ax.set_ylabel("Frequency")

plt.tight_layout()
plt.show()


                         
            
```

```python
#correlation between Sales and profit
#We can deduce that there is no form of correlation
sales = data['Sales'].head(20)
profit = data['Profit'].head(20)
plt.scatter(sales, profit)
plt.title("Correlation of Sales & Profit")
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.show()
```

```python
#Adding a new column
df = pd.DataFrame(data)
production_cost = df["Sales"] - df["Profit"]
df["Production Cost"] = production_cost
print (df)
```

```python
#correlation between production cost and sales
# There is a +ve correlation
sales = data['Sales'].head(20)
production_cost = df['Production Cost'].head(20)
plt.scatter(sales, production_cost)
plt.title("Correlation of Sales & Production Cost")
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.show()
```
